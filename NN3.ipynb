```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
# Set the random seed for reproducibility
np.random.seed(42)
```

```python
# Define the input data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
```

```python
# Define the neural network architecture
input_size = 2
hidden_size = 4
output_size = 1
```

```python
# Define the learning rate and number of training epochs
learning_rate = 0.1
num_epochs = 10000
```

```python
# Define the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))
```

```python
# Convert input and output numpy arrays to PyTorch tensors
X_tensor = torch.from_numpy(X).float()
y_tensor = torch.from_numpy(y).float().view(-1, 1)
```

```python
# Define weights and biases
W1 = torch.randn(input_size, hidden_size, requires_grad=True)
b1 = torch.randn(1, hidden_size, requires_grad=True)
W2 = torch.randn(hidden_size, output_size, requires_grad=True)
b2 = torch.randn(1, output_size, requires_grad=True)
```

```python
# Train the network using gradient descent
losses = []
for epoch in range(num_epochs):
    # Forward pass
    z1 = torch.matmul(X_tensor, W1) + b1
    a1 = sigmoid(z1)
    z2 = torch.matmul(a1, W2) + b2
    y_pred = sigmoid(z2)
    
    # Compute and print loss
    loss = torch.nn.functional.binary_cross_entropy(y_pred, y_tensor)
    print(f'Epoch {epoch + 1} loss: {loss.item():.4f}')
    losses.append(loss.item())
    
    # Backward pass
    loss.backward()
    
    # Update the weights and biases using gradient descent
    with torch.no_grad():
        W1 -= learning_rate * W1.grad
        b1 -= learning_rate * b1.grad
        W2 -= learning_rate * W2.grad
        b2 -= learning_rate * b2.grad
        
        # Reset the gradients to zero
        W1.grad.zero_()
        b1.grad.zero_()
        W2.grad.zero_()
        b2.grad.zero_()
```

```python
# Plot the training loss over time
plt.plot(losses)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
# Use the trained network to make predictions on new data
with torch.no_grad():
    z1 = torch.matmul(X_tensor, W1) + b1
    a1 = sigmoid(z1)
    z2 = torch.matmul(a1, W2) + b2
    y_pred = sigmoid(z2)
    print("Predictions:", y_pred.numpy())
```
